```python
stores = pd.read_csv('Data/stores.csv')
oil = pd.read_csv('Data/oil.csv')
transactions =  pd.read_csv('Data/transactions.csv')
holidays_events = pd.read_csv('Data/holidays_events.csv')
```

A continuación, lo que vamos a hacer va a ser combinar los diferentes DataFrames de datos de los que disponemos con el set de entrenamiento `train`. En primer lugar, hacemos la combinación con `stores`, con respecto a la variable `store_nbr`. El resultado es el siguiente: 

```python
train = train.merge(stores, how="outer", on='store_nbr')            
```

```python
train
```

```python
oil.head()
```

```python
train = train.merge(oil, how="outer", on='date')            
```

```python
train[['store_nbr','onpromotion','cluster']] = train[['store_nbr','onpromotion','cluster']].astype(int)
```

Si nos fijamos en el DataFrame anterior, vemos como disponemos de datos del aceite para fechas en las que no tenemos registrados datos del set de entrenamiento. Entonces procedemos a eliminar dichas filas. Para ello vamos a borrar aquellas filas cuya fecha sea posterior al 2017-08-15. 

```python
borrar = range(3000888,3000903)
train = train.drop(borrar,axis=0)
```
